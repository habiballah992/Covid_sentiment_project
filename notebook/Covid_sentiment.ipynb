## 3-cleaning and processing the data

### change case of text and rmove the spaces from begin and end

```python
df['clean_tweet']=df['clean_tweet'].astype(str).str.lower().str.strip()
```

### drop duplicates and null values

```python
df.dropna(inplace=True)
```

```python
df=df.drop_duplicates(subset='clean_tweet')
```

### drop the empty String

```python
df=df[df['clean_tweet'] !="" ]
```

```python
df['sentiment']=['positive' if s=='pos' else 'negative' if s=='neg' else 'natural' for s in df['sentiment']]
```

## 4-split and vectorize the data

```python
x=df['clean_tweet']
y=df['sentiment']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
vectorizer=CountVectorizer()

x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
```

## 5-training modeles(SVM,RandomForest)

```python
Forest_model=RandomForestClassifier()
Forest_model.fit(x_train,y_train)
```

```python
svm_model=LinearSVC()
svm_model.fit(x_train,y_train)
```
